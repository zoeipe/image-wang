```python
from matplotlib import pyplot as p
import numpy as np
import scipy.ndimage

I = np.random.rand(256,256)

kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
I_sharp = scipy.ndimage.convolve(I, kernel)

p.imshow(I_sharp, cmap="gray")
p.show()
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('girl.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
blurred = cv2.GaussianBlur(img_rgb, (11, 11), 0)

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("picture")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(blurred)
plt.title("after picture")
plt.axis('off')

plt.show()
```

```python
kernel = np.ones((5, 5), np.float32) / 25
blurred = cv2.filter2D(img_rgb, -1, kernel)

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("picture")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(blurred)
plt.title("after picture")
plt.axis('off')

plt.show()
```

```python
kernel = np.array([
    [-1, -1, -1],
    [-1,  9, -1],
    [-1, -1, -1]
])

blurred = cv2.filter2D(img_rgb, -1, kernel)

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("picture")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(blurred)
plt.title("after picture")
plt.axis('off')

plt.show()
```

```python
sharpen_kernel = np.array([
    [ 0, -1,  0],
    [ -1,  5, -1],
    [ 0, -1,  0]
])

sharpened_img = cv2.filter2D(img_rgb, -1, sharpen_kernel)

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("picture")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sharpened_img)
plt.title("after picture")
plt.axis('off')

plt.show()
```

```python
sharpen_kernel = np.array([
    [ 0 , -1, 0],
    [ -1, 5, -1],
    [ 0, -1, 0]
])

sharpened_img1 = cv2.filter2D(img_rgb, -1, sharpen_kernel)
sharpened_img2 = cv2.filter2D(sharpened_img1, -1, sharpen_kernel)

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("picture")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sharpened_img2)
plt.title("after picture")
plt.axis('off')

plt.show()
```
